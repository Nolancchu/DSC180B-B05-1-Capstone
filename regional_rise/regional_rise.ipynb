```python
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import numpy as np
```

```python
# Load reference scenario
ds_ref = xr.open_dataset("slr_ssp126.nc")
slr_ref = ds_ref["slr"].sel(time=2100) - ds_ref["slr"].sel(time=2015)

# Load high-emissions scenario
ds_hi = xr.open_dataset("slr_ssp126.nc")
slr_hi = ds_hi["slr"].sel(time=2100) - ds_hi["slr"].sel(time=2015)

# Difference
slr_diff = slr_hi - slr_ref

lon = ds_ref["lon"].values
lat = ds_ref["lat"].values
LON, LAT = np.meshgrid(lon, lat)

fig = plt.figure(figsize=(12, 6))
ax = plt.axes(projection=ccrs.Robinson())

pcm = ax.pcolormesh(
    LON,
    LAT,
    slr_diff.values,
    transform=ccrs.PlateCarree(),
    cmap="RdBu_r",
    vmin=-0.6,
    vmax=0.6
)

ax.coastlines()
ax.set_title("Additional regional sea-level rise:\nSSP1-2.6 minus SSP1-2.6 (2100  2015)")

cbar = plt.colorbar(pcm, orientation="horizontal", pad=0.05)
cbar.set_label("Additional sea-level rise (m)")

plt.show()
```

```python
import matplotlib.colors as colors

ds = xr.open_dataset("slr_ssp126.nc")
slr_delta = ds["slr"].sel(time=2100) - ds["slr"].sel(time=2015)

lon = ds["lon"].values
lat = ds["lat"].values
LON, LAT = np.meshgrid(lon, lat)

fig = plt.figure(figsize=(12, 6))
ax = plt.axes(projection=ccrs.Robinson())

norm = colors.LogNorm(
    vmin=0.01,                              # 1 cm
    vmax=np.nanpercentile(slr_delta, 99)   # robust upper bound
)

pcm = ax.pcolormesh(
    LON,
    LAT,
    slr_delta.values,
    transform=ccrs.PlateCarree(),
    cmap="inferno",
    norm=norm
)

ax.coastlines()
ax.set_title("Regional sea-level rise (log scale)\nSSP1-2.6, 2100  2015")

cbar = plt.colorbar(pcm, orientation="horizontal", pad=0.05)
cbar.set_label("Sea-level rise (m, log scale)")

plt.show()
```
